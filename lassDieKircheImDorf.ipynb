```python
def draw_blue_board_with_numbers(cell_size=6):
    size = 7  # Fix auf 7x7 – kann nicht verändert werden

    BLUE = "\033[44m"
    RESET = "\033[0m"

    print("\n        Sehr gut sichtbares 7×7-Spielfeld (blau)\n")

    # Spaltennummern - LINKSBÜNDIG für bessere Ausrichtung
    print("     ", end="")
    for c in range(size):
        # Linksbündige Formatierung: < statt ^
        print(f"{c:<{cell_size}}", end="")
    print()

    # obere Rahmenlinie
    print("    " + "┌" + ("─" * cell_size + "┬") * (size-1) + "─" * cell_size + "┐")

    for r in range(size):
        # Zeilennummer - immer gut sichtbar
        print(f"{r:>3} │", end="")
        
        for c in range(size):
            print(BLUE + " " * cell_size + RESET + "│", end="")
        print()

        # Zwischenlinien außer nach letzter Zeile
        if r < size - 1:
            print("    " + "├" + ("─" * cell_size + "┼") * (size-1) + "─" * cell_size + "┤")

    # untere Rahmenlinie
    print("    " + "└" + ("─" * cell_size + "┴") * (size-1) + "─" * cell_size + "┘")

draw_blue_board_with_numbers()
```
